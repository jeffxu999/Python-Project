```
# DNA Replication & Visualization - Python Project

###################################################################

# 1) randomly generate DNA

# 2) create complementary DNA strand (DNA replication)

# 3) count nucleotides (A vs T vs C vs G)

# 4) print out double helix 

# 5) graph out to visualize

###################################################################

import numpy as np
import matplotlib.pyplot as plt

###################################################################

def create_dna(nucleotide_length):
    
    our_dna = ""

    for index in range(nucleotide_length):
        our_nucleotide = np.random.randint(1,5)
        if our_nucleotide == 1:
            our_dna += "A"
        elif our_nucleotide == 2:
            our_dna += "T"
        elif our_nucleotide == 3:
            our_dna += "C"
        else:
            our_dna += "G"

    return our_dna


coding_strand = create_dna(999)

# print(coding_strand)

# print(len(coding_strand))

###################################################################

def create_complementary_dna(five_to_three):
    
    our_dna = ""

    for nucleotide in five_to_three:
        if nucleotide == "A":
            our_dna += "T"
        elif nucleotide == "T":
            our_dna += "A"
        elif nucleotide == "C":
            our_dna += "G"
        else:
            our_dna += "C"

    return our_dna


template_strand = create_complementary_dna(coding_strand)

# print(template_strand)

# print(len(template_strand))

###################################################################

def count_nucleotides(some_dna):
    As = 0
    Ts = 0
    Cs = 0
    Gs = 0
    
    for nucleotide in some_dna:
        if nucleotide == "A":
            As += 1
        elif nucleotide == "T":
            Ts += 1
        elif nucleotide == "C":
            Cs += 1
        else:
            Gs += 1

    return As, Ts, Cs, Gs


A_count, T_count,C_count, G_count = count_nucleotides(coding_strand)

# print("A: ",A_count)
# print("T: ",T_count)
# print("C: ",C_count)
# print("G: ",G_count)

###################################################################

def double_helix(five_to_three):
    print("5'",five_to_three,"3'")
    print("3'",create_complementary_dna(five_to_three),"5'")
    
    
# double_helix(coding_strand)

###################################################################

def visualize_dna(some_dna):
    length = np.ceil(np.sqrt(len(some_dna)))
    
    x = 1
    y = length+1
    
    for index in range(int(length*length)):
        color="pink"
        
        if index%length == 0:
            y -= 1
            x = 1
        else:
            x += 1
            
        try:
            nucleotide = some_dna[index]
            if nucleotide == "A":
                color = "red"
            elif nucleotide == "T":
                color = "lime"
            elif nucleotide == "C":
                color = "orange"
            else:
                color = "purple"
        except IndexError:
            color="black"
        
        plt.scatter(x,y,c=color,marker="s")
        
    plt.show()
    
    
print(coding_strand)
    
visualize_dna(coding_strand)
```
